```python
# read the CSV file into a pandas DataFrame
dil = pd.read_csv('driver_imgs_list.csv')
```

```python
dil.head()
```

```python
# group the frame by the subject in the image
grouped = dil.groupby('subject')
```

```python
grouped.get_group('p002').head()
```

```python
grouped.groups.keys()
```

```python
# pick 3 subjects at random (I'd suggest a validation set with more drivers than that however - I'm using 3)
subjects = np.random.permutation(grouped.groups.keys())[:3]
```

```python
print(subjects)
```

```python
%pwd
```

```python
# get the group associated with the subject
for subject in subjects:
    group = grouped.get_group(subject)
    # loop over the group and move the images into the validation directory
    for (subject, cls, img) in group.values:
        source = 'train/{}/{}'.format(cls, img)
        target = source.replace('train', 'valid')
        print('mv {} {} {}'.format(subject, source, target))
        os.rename(source, target)
```
